```python
import pandas as pd

df = pd.read_excel('gastos2.xlsx')
```

```python
df = df.groupby(['Date', 'Cuenta'],as_index=False)
df1= df[['Precio']].sum()
df1[:10]
```

```python
per = df1.Date.dt.to_period("M")
g = df1.groupby(per)
g.sum()
```

```python
df_grouped = df.loc[:, ["Date", "Cuenta", "Precio"]]
df_grouped = df_grouped.groupby(["Date", "Cuenta"])
df_grouped
for key, item in df_grouped:
    group = df_grouped.get_group(key)
    print(group, "\n")
```
